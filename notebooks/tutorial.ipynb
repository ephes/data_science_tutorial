# Machine Learning mit Python

Machine Learning als Fachgebiet erhält in den letzten Jahren zunehmend Aufmerksamkeit.
Es bildet die Grundlage für viele der interessantesten Produkte und Dienstleistungen,
die im IT-Umfeld heutzutage vorgestellt werden.

Ein Grossteil der Machine-Learning Systeme wird in Python entwickelt, weil viele der
beliebtesten Softwarebibliotheken entweder direkt in Python geschrieben wurden oder
zumindest über eine Python-Anbindung verfügen.

Nutzbare Softwareinfrastruktur ist aber nur eine der Vorraussetzungen, um erfolgreich
Machine-Learning Projekte umsetzen zu können. Eine andere sind Daten. Es ist sicher
schwer vorstellbar, im gleichen Umfang wie Google Daten erheben und verarbeiten zu
können. Andererseits ist das aber auch nicht immer erforderlich. Es gibt überraschend
viele Machine Learning Anwendungen, die sich mit handhabbareren Datenmengen umsetzen
lassen.

Dieser Kurs bietet einen Überblick über das PyData-Ökosystem und einen Einblick in
Best Practices und Tools zum Bau von Machine-Learning Systemen in Python. Behandelt
werden beispielsweise:

- Jupyter Notebooks
- Numpy/Pandas/Matplotlib
- Scikit-Learn
- Keras/Tensorflow
